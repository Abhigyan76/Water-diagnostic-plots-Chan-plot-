```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
start_date = datetime(1996, 1, 1)
end_date = datetime(2016, 12, 1)
dates = pd.date_range(start=start_date, end=end_date, freq='MS')

wor_values = np.random.uniform(2, 10, size=len(dates))

rate_change_wor = np.random.uniform(2, 10, size=len(dates))

data = pd.DataFrame({
    'Date': dates,
    'WOR': wor_values,
    'Rate_of_change_WOR': rate_change_wor
})

rapid_syn = data[(data['WOR'] >= 2) & (data['Rate_of_change_WOR'] > 2)].head(600)

display(rapid_syn)
```

```python
start_date = datetime(1996, 1, 1)
end_date = datetime(2018, 12, 1)
dates = pd.date_range(start=start_date, end=end_date, freq='MS')

wor_values = np.random.uniform(0, 1, size=len(dates))

rate_change_wor = np.random.uniform(1, 2, size=len(dates))

data3 = pd.DataFrame({
    'Date': dates,
    'WOR': wor_values,
    'Rate_of_change_WOR': rate_change_wor
})

const_syn = data3[(data3['WOR'] < 1) & (data3['Rate_of_change_WOR'] > 1) & (data3['Rate_of_change_WOR'] < 2)].head(600)

display(const_syn)
```

```python

start_date = datetime(1996, 1, 1)
end_date = datetime(2028, 12, 1)
dates = pd.date_range(start=start_date, end=end_date, freq='MS')


wor_values = np.random.uniform(0.9, 1.1, size=len(dates))

rate_change_wor = np.random.uniform(3.1, 4, size=len(dates))
rate_change_wor = np.where(np.random.rand(len(dates)) > 0.5, rate_change_wor, -rate_change_wor)

data4 = pd.DataFrame({
    'Date': dates,
    'WOR': wor_values,
    'Rate_of_change_WOR': rate_change_wor
})

normal_syn = data4[(data4['WOR'] <= 1) & (data4['Rate_of_change_WOR'] > 3) & (data4['Rate_of_change_WOR'] < 4)].head(600)

display(normal_syn)
```

```python
rapid_syn.to_csv('rapid_syn.csv', index=False)
```
